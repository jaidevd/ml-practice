```python
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier
from sklearn import model_selection as cv

wine = load_wine()
xtr, xts, ytr, yts = cv.train_test_split(wine.data, wine.target, test_size=0.3, random_state=1)

dt = DecisionTreeClassifier(random_state=1).fit(xtr, ytr)
```

```python
dt.score(xtr, ytr)
```

```python
dt.score(xts, yts)
```

```python
dt.get_depth()
```

```python
dt = DecisionTreeClassifier(
    random_state=1,
    criterion='entropy',
    splitter='random',
    max_depth=3,
    min_samples_split=4,
    min_impurity_decrease=0.1
).fit(xtr, ytr)
```

```python
dt.score(xtr, ytr)
```

```python
dt.score(xts, yts)
```

```python
dt.tree_.impurity
```
